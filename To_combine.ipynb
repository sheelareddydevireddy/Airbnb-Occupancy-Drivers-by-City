```python
import os
import pandas as pd

master_df = pd.DataFrame()

for file in os.listdir(os.getcwd()): 
    if file.endswith(".csv"):
        df = pd.read_csv(file)
        df["State"] = file.replace(".csv", "")
        master_df = pd.concat([master_df, df], ignore_index=True)

master_df.to_csv("Master File State.CSV", index = False)
```
